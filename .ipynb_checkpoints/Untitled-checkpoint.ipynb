```python
%matplotlib inline
from matplotlib.pylab import *
import seaborn as sns
import numpy as np
from collections import defaultdict
import random


font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 32}

matplotlib.rc('font', **font)

sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
data_a = np.random.normal(1,size = (400000,))
data_b = np.random.normal(1,size = (400000,))
```

```python
figure(figsize = (16,10))
plot(data_a+0.005,ls = '',marker = 'o')
plot(data_b-0.005,ls = '',marker = 'o')
xlim(right = 100000)
xlabel("sample")
ylabel("data value")
```

```python
avg_a = np.array([np.mean(data_a[:i]) for i in range(len(data_a))])
std_a = np.array([np.std(data_a[:i])/np.sqrt(i) for i in range(1,len(data_a))])
avg_b = np.array([np.mean(data_b[:i]) for i in range(len(data_b))])
std_b = np.array([np.std(data_b[:i])/np.sqrt(i) for i in range(1,len(data_b))])
```

```python
import matplotlib.gridspec as gridspec
figure(figsize = (16,10))

gs1 = gridspec.GridSpec(2, 1)
gs1.update(wspace=0.025, hspace=0.05) # set the spacing between axes. 

frame = subplot(gs1[0])
plot(data_a+0.005,ls = '',marker = 'o')
plot(data_b-0.005,ls = '',marker = 'o')
xlim(right = 100000)
yticks([-4,-2,0,2,4])
ylim(-4)
frame.axes.get_xaxis().set_visible(False)
ylabel("raw signals")
subplot(gs1[1])
plot(avg_a+0.005,ls = '',marker = '.')
plot(avg_b-0.005,ls = '',marker = '.')
plot(avg_a[1:]+std_a+0.005,lw = 0.2,ls = '--')
plot(avg_a[1:]-std_a+0.005,lw = 0.2,ls = '--')
plot(avg_b[1:]+std_b-0.005,lw = 0.2,ls = '--')
plot(avg_b[1:]-std_b-0.005,lw = 0.2,ls = '--')
axhline(1.005,ls = '--',color = 'grey',lw = 1)
axhline(0.995,ls = '--',color = 'grey',lw = 1)
ylim(0.98,1.02)
yticks([0.98,0.99,1.0,1.01])
xlim(right = 200000)
xlabel("# data points")
ylabel("average signals, standard error")
```
