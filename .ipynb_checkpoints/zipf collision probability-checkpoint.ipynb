```python
%matplotlib inline
from matplotlib.pylab import *
import seaborn as sns
import numpy as np
```

```python
import math
def p_collision(p0,n):
    p = 1.0
    for i in range(1,n+1):
        pk = p0/i
        p *= 1-2*pk*(1-pk)
    return p
```

```python
p_collision(0.5,1000)
```

```python
xs = np.linspace(0,1,1000)
for i in np.logspace(1.,3.,10):
    print i
    ps = np.array([p_collision(x,int(i)) for x in xs])
    plot(xs,ps)
semilogy()
```
